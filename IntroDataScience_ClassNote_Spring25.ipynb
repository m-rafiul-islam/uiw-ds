**Mutable:** You can change the contents of a list after it's been created, such as adding, removing, or modifying elements.

**Ordered:** The elements in a list have a defined order, and that order will not change unless you explicitly change it. This allows indexing and slicing operations.

**Versatile:** Lists can hold any type of data, and the elements within a list don't need to be of the same type. You can have a list containing integers, floats, strings, and other lists all at once.

## What are Control Flows?
Control flows are the decision-making structures in programming that determine the execution path of a program based on conditions or loops. They direct the computer on when to start, stop, or skip specific instructions based on logical decisions or repetitive conditions.

## Key Concepts in Control Flows

### 1. Conditional Statements
   - **If Statements**: Execute a block of code only if a specified condition is true.
   - **Else Statements**: Execute a block of code if the condition in the if statement is false.
   - **Elif Statements**: Specify a new condition to test if the previous conditions were false.

### 2. Looping Statements
   - **For Loops**: Iterate over a sequence (like a list, tuple, or string) and execute a block of code once for each item.
   - **While Loops**: Continuously execute a block of code as long as a condition is true.

### 3. Control Statements
   - **Break**: Exit the loop entirely.
   - **Continue**: Skip the rest of the code inside the loop for the current iteration and move to the next iteration.
   - **Pass**: A placeholder statement, syntactically needed, but performs no operation.

The following is a table of the repetition markers available for use in regular expressions:

| Character | Description | Example |
|-----------|-------------|---------|
| ``?`` | Match zero or one repetitions of preceding  | ``"ab?"`` matches ``"a"`` or ``"ab"`` |
| ``*`` | Match zero or more repetitions of preceding | ``"ab*"`` matches ``"a"``, ``"ab"``, ``"abb"``, ``"abbb"``... |
| ``+`` | Match one or more repetitions of preceding  | ``"ab+"`` matches ``"ab"``, ``"abb"``, ``"abbb"``... but not ``"a"`` |
| ``{n}`` | Match ``n`` repetitions of preeeding | ``"ab{2}"`` matches ``"abb"`` |
| ``{m,n}`` | Match between ``m`` and ``n`` repetitions of preceding | ``"ab{2,3}"`` matches ``"abb"`` or ``"abbb"`` |

# Numpy
NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays. It is one of the core packages for scientific computing with Python and is widely used in machine learning, data science, and scientific computing communities.

NumPy provides efficient and convenient support for array computation, and its multi-dimensional arrays provide a structure for organizing and processing large amounts of homogeneous data, making it a powerful tool for solving a wide range of computational problems in science and engineering.



## Why numpy is important?

NumPy is a library for the Python programming language that provides support for arrays and matrices, along with a collection of mathematical functions to operate on these arrays. It is widely used for numerical computing, data analysis, and scientific computing.

There are several reasons why NumPy is important:

- Efficient Computation: NumPy arrays are much more efficient for numerical computations compared to lists in Python, as the data is stored in a contiguous block of memory and can be processed much more quickly by the computer.

- Broadcasting: NumPy supports broadcasting, which allows for operations to be performed between arrays of different shapes, by repeating the smaller array to match the shape of the larger array.

- Integration with Other Libraries: NumPy provides support for reading and writing arrays to and from files, as well as integrating with other libraries, such as Matplotlib for plotting and visualization, and Pandas for data analysis.

- Element-wise Operations: One of the key features of NumPy is its ability to perform element-wise operations on arrays, allowing for fast and efficient computation.

## Book store

https://books.toscrape.com/

````python
import requests
from bs4 import BeautifulSoup as bs
````

````python
# single page
url  = 'https://books.toscrape.com/catalogue/page-1.html'
````

````python
def url_to_soup(url):
  response = requests.get(url)
  html_content = response.content
  soup = bs(html_content)
  return soup

def get_book_title(soup):
  h3_tag_list = soup.find_all('h3')
  return [item.a['title'] for item in h3_tag_list]

def get_short_url(soup):
  h3_tag_list = soup.find_all('h3')
  return [item.a['href'] for item in h3_tag_list]

def get_book_price(soup):
  price_tag_list = soup.find_all('p',class_='price_color')
  return [item.text for item in price_tag_list]
````

<p class="price_color">£51.77</p>

````python
soup = url_to_soup(url)
get_short_url(soup)
````

````python
title, short_url, price = [], [], []
for i in range(1,51):
  url  = f'https://books.toscrape.com/catalogue/page-{i}.html'
  soup = url_to_soup(url)
  title.extend(get_book_title(soup))
  short_url.extend(get_short_url(soup))
  price.extend(get_book_price(soup))
````

````python
book_dictionary = {'Title':title,
                   'Short Link':short_url,
                   'Book Price': price}
````

````python
import pandas as pd
df = pd.DataFrame(book_dictionary)
df.to_csv('book_store.csv')
df
````
