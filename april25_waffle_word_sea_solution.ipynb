# Waffle Chart

What is a waffle charts? 

- A waffle chart is a great way to visualize data in relation to a whole or to highlight progress against a given threshold. 


```python
# pip install pywaffle 
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

```python
# Vote Percentage in 2016 US Presidential Election
data = {'Democratic': 48, 'Republican': 46, 'Libertarian': 3} 
fig = plt.figure(
    FigureClass=Waffle, 
    rows=5, 
    values=data
)
```

```python
data = {'Democratic': 48, 'Republican': 46, 'Libertarian': 3}
fig = plt.figure(
    FigureClass=Waffle, 
    rows=5, 
    values=data, 
    colors = ("#00AEF3", "#DE0100", "#FED105"),
    title={'label': 'Vote Percentage in 2016 US Presidential Election', 'loc': 'left'},
    labels=["{0} ({1}%)".format(k, v) for k, v in data.items()],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': len(data)}
)
```

```python
fig = plt.figure(
    FigureClass=Waffle, 
    rows=5, 
    values=data, 
    colors = ("#00AEF3", "#DE0100", "#FED105"),
    title={'label': 'Vote Percentage in 2016 US Presidential Election', 'loc': 'left'},
    labels=["{0} ({1}%)".format(k, v) for k, v in data.items()],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': len(data)},
    icons='child', icon_size=18, 
    icon_legend=True
    
)
```

# Word Clouds

**What is a Word Cloud?**

A word cloud (also known as a tag cloud) is a visual representation of words. 

A word cloud is simply a depiction of the importance of different words in the body of text. A word cloud works in a simple way; the more a specific word appears in a source of textual data the bigger and bolder it appears in the world cloud.


```python
# pip install wikipedia
# conda install -c conda-forge wordcloud
# pip install worldcloud 
```

```python
import sys
import wikipedia
from wordcloud import WordCloud, STOPWORDS
```

```python
def get_wiki(query):
	# get best matching title for given query
	title = wikipedia.search(query)[0]
	# get wikipedia page for selected title
	page = wikipedia.page(title)
	return page.content
```

```python
#get the text from wikipedia 
text = get_wiki('data science') 
# get_wiki('python programming language')

# instantiate a word cloud object
wc = WordCloud()

# generate the word cloud
wc.generate(text)

# save the word cloud into a file 
wc.to_file('python_wc.png')
# display the word cloud
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```

```python
stopwords = set(STOPWORDS)
# stopwords
wc = WordCloud(background_color='white',max_words = 2000, stopwords = stopwords)
wc.generate(text)
wc.to_file('python_wc_2.png')
# display the word cloud
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
df.to_csv('tips.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=df)
```

```python
#markers
sns.lmplot(x='total_bill',y='tip',hue='sex',markers=['o','*'],data=df)
```

```python
# grid 
sns.lmplot(x='total_bill',y='tip',col='sex',data=df)
```

```python
# grids 
sns.lmplot(x='total_bill',y='tip',col='sex',row='time',data=df)
```
